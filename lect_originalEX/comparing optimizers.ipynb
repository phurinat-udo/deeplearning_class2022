```python
from tensorflow.keras import Sequential, layers
```

```python
model = Sequential()
model.add(layers.Dense(1, input_shape=(1,)))
model.summary()
```

```python
import numpy as np
from matplotlib import pyplot as plt
n = 50
x = np.linspace(0, 1, n)
w = 5; b = -4 # true parameters
y = w*x + b + np.random.randn(n)
plt.plot(x,y,'o')
plt.show()
```

```python
model.compile(loss='mse', optimizer='sgd')
```

```python
history = model.fit(x,y,epochs=2,batch_size=1)
```

```python
plt.plot(history.epoch, history.history['loss'])
```

```python
# SGD
model = Sequential()
model.add(layers.Dense(1, input_shape=(1,)))
model.compile(loss='mse', optimizer='sgd')
h_sgd = model.fit(x,y,epochs=1000,batch_size=1)
```

```python
# BGD
model = Sequential()
model.add(layers.Dense(1, input_shape=(1,)))
model.compile(loss='mse', optimizer='sgd')
h_bgd = model.fit(x,y,epochs=1000,batch_size=50)
```

```python
# MGD
model = Sequential()
model.add(layers.Dense(1, input_shape=(1,)))
model.compile(loss='mse', optimizer='sgd')
h_mgd = model.fit(x,y,epochs=1000,batch_size=10)
```

```python
from tensorflow.keras import optimizers
# BGD with Classical momentum
model = Sequential()
model.add(layers.Dense(1, input_shape=(1,)))
sgd = optimizers.SGD(momentum=0.9)
model.compile(loss='mse', optimizer=sgd)
h_bgd_cm = model.fit(x,y,epochs=1000,batch_size=50)
```

```python
# BGD with NAG
model = Sequential()
model.add(layers.Dense(1, input_shape=(1,)))
sgd = optimizers.SGD(momentum=0.9, nesterov=True)
model.compile(loss='mse', optimizer=sgd)
h_bgd_nag = model.fit(x,y,epochs=1000,batch_size=50)
```

```python
plt.subplot(121)
plt.plot(h_bgd.epoch,h_bgd.history['loss'],'b')
plt.plot(h_sgd.epoch,h_sgd.history['loss'],'r')
plt.plot(h_mgd.epoch,h_mgd.history['loss'],'k')
plt.legend(['BGD','SGD','MGD10'])
plt.subplot(122)
plt.plot(h_bgd.epoch,h_bgd.history['loss'],'b')
plt.plot(h_bgd.epoch,h_bgd_cm.history['loss'],'r')
plt.plot(h_bgd.epoch,h_bgd_nag.history['loss'],'k')
plt.legend(['BGD','BGD+CM0.9','BGD+NAG0.9'])
plt.show()
```

```python
# AdaGrad
#tf.keras.optimizers.Adagrad(learning_rate=0.001,
#                            initial_accumulator_value=0.1,
#                            epsilon=1e-07)
model = Sequential()
model.add(layers.Dense(1, input_shape=(1,)))
op = optimizers.Adagrad()
model.compile(loss='mse', optimizer=op)
h_adagrad = model.fit(x,y,epochs=1000,batch_size=50)
```

```python
# AdaDelta
# Adadelta(learning_rate=0.001,rho=0.95,epsilon=1e-07)
model = Sequential()
model.add(layers.Dense(1, input_shape=(1,)))
op = optimizers.Adadelta()
model.compile(loss='mse', optimizer=op)
h_adadelta = model.fit(x,y,epochs=1000,batch_size=50)
```

```python
# RMSprop
#RMSprop(learning_rate=0.001,rho=0.9,momentum=0.0,epsilon=1e-07,centered=False)
model = Sequential()
model.add(layers.Dense(1, input_shape=(1,)))
op = optimizers.RMSprop()
model.compile(loss='mse', optimizer=op)
h_rmsprop = model.fit(x,y,epochs=1000,batch_size=50)
```

```python
# RMSprop with momentum
# RMSprop
#RMSprop(learning_rate=0.001,rho=0.9,momentum=0.0,epsilon=1e-07,centered=False)
model = Sequential()
model.add(layers.Dense(1, input_shape=(1,)))
op = optimizers.RMSprop(momentum=0.9)
model.compile(loss='mse', optimizer=op)
h_rmsprop_cm = model.fit(x,y,epochs=1000,batch_size=50)
```

```python
# Adam
#Adam(learning_rate=0.001,beta_1=0.9,beta_2=0.999,epsilon=1e-07,amsgrad=False)
model = Sequential()
model.add(layers.Dense(1, input_shape=(1,)))
op = optimizers.Adam()
model.compile(loss='mse', optimizer=op)
h_adam = model.fit(x,y,epochs=1000,batch_size=50)
```

```python
# AMSgrad
#Adam(learning_rate=0.001,beta_1=0.9,beta_2=0.999,epsilon=1e-07,amsgrad=False)
model = Sequential()
model.add(layers.Dense(1, input_shape=(1,)))
op = optimizers.Adam(amsgrad=True)
model.compile(loss='mse', optimizer=op)
h_amsgrad = model.fit(x,y,epochs=1000,batch_size=50)
```

```python
# AdaMax
#AdaMax(learning_rate=0.001,beta_1=0.9,beta_2=0.999,epsilon=1e-07)
model = Sequential()
model.add(layers.Dense(1, input_shape=(1,)))
op = optimizers.Adamax()
model.compile(loss='mse', optimizer=op)
h_adamax = model.fit(x,y,epochs=1000,batch_size=50)
```

```python
# Nadam
#Nadam(learning_rate=0.001,beta_1=0.9,beta_2=0.999,epsilon=1e-07)
model = Sequential()
model.add(layers.Dense(1, input_shape=(1,)))
op = optimizers.Nadam()
model.compile(loss='mse', optimizer=op)
h_nadam = model.fit(x,y,epochs=1000,batch_size=50)
```

```python
epoch = h_bgd.epoch
plt.plot(epoch,h_adagrad.history['loss'],'b')
plt.plot(epoch,h_adadelta.history['loss'],'r')
plt.plot(epoch,h_rmsprop.history['loss'],'k')
plt.plot(epoch,h_rmsprop_cm.history['loss'],'k--')
plt.plot(epoch,h_adam.history['loss'],'b--')
plt.plot(epoch,h_amsgrad.history['loss'],'r--')
plt.plot(epoch,h_nadam.history['loss'],'b:')
plt.legend(['Adagrad','Adadelta','RMSprop','RMSprop+CM','Adam','AMSgrad','Nadam'])
plt.show()
```
